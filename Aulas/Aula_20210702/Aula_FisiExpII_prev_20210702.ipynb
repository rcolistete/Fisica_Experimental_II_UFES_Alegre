# SciDAVis

Os gráficos dos relatórios dessa disciplina devem ser feitos usando o software SciDAVis, ou via Python + certos módulos (opcional).

O [SciDAVis](http://scidavis.sourceforge.net/) é uma continuação desde 2007 do QtiPlot (que por sua vez é inspirado no Origin que é comercial), vide [histórico](https://en.wikipedia.org/wiki/SciDAVis).

O nome SciDAVis vem "application for **Sci**entific **D**ata **A**nalysis and **Vis**ualization", ou seja, SciDAVis é uma aplicação/software para "Análise e Visualização de Dados Científicos".

O SciDAVis (Scientific Data Analysis and Visualization) é gratuito e open-source, funciona em PC (Linux, Mac, Windows), faz análise estatística e gráfica de dados experimentais/observacionais.

Passo-a-passo com exemplo simples de uso de SciDAVis para fazer gráfico de v(t) x t usando dados com incertezas, com ajuste de curva linear :

- digitar os dados do eixo horizontal (1[X]) e vertical (2[Y]), sem incertezas;
- clicando com o botão direito do mouse na tabela (ou menu Table), selecioonar "Add Column" duas vezes;
- clicando na 3a coluna, com o botão direito do mouse escolher "Set Column(s) As->X Error", aí será renomeada para (3[xEr]). Digite as incertezas do eixo horizontal nessa coluna;
- clicando na 4a coluna, com o botão direito do mouse escolher "Set Column(s) As->Y Error", aí será renomeada para (3[yEr]). Digite as incertezas do eixo vertical nessa coluna;
- selecione as 4 colunas, então via menu ou botão direito do mouse, escolha "Plot -> Scatter";
- depois selecione menu "Analysis -> Quick Fit -> Fit Linear";
- clicar em cada eixo e título do gráfico para renomear;
- clique em um eixo do gráfico, depois escolhe "Grid", selecione "Major Grids" em "Horizontal" e "Vertical";
- opcional : clique na legenda no canto superior esquerdo, escola "Delete" para apagá-la.
