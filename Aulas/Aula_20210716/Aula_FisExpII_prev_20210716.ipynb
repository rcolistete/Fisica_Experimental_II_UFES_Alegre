### [LTSpice](https://www.analog.com/en/design-center/design-tools-and-calculators/ltspice-simulator.html)

![image.png](attachment:image.png "LTSpice")

[LTSpice](https://www.analog.com/en/design-center/design-tools-and-calculators/ltspice-simulator.html) é um software gratuito de computador, simulador de circuito eletrônico analógico baseado em [SPICE](http://bwrcs.eecs.berkeley.edu/Classes/IcBook/SPICE/), produzido pelo fabricante de semicondutores [Analog Devices](https://www.analog.com/en/index.html).

[SPICE](http://bwrcs.eecs.berkeley.edu/Classes/IcBook/SPICE/) é o software mais amplamente distribuído e usado na indústria para simulação de circuitos eletrônicos, sendo referência mundial.

LTSpice em versões gratuitas para Mac OS e Windows. No Linux, LTSpice roda via WINE (Wine Is Not an Emulator).

LTSpice tem muitos recursos, exemplos e tutoriais, devido ao suporte oficial da Analog Devices e herança do SPICE. Porém é mais complexo de aprender.

### [Fritzing](https://fritzing.org/download/)

![image.png](attachment:image.png "Fritzing")

[Fritzing](https://fritzing.org/download/) é um software de [código-fonte aberto](https://github.com/fritzing/) e gratuito de computador para EDA (Electronic Design Automation), sendo muito amigável e popular. **Tem modo protoboard que é muito fácil de usar**, só arrastar os componentes, com uma biblioteca gigantesca. E a interface tem versão em português (pt-br).

Embora para o download seja pedida doação (EUR8/25), Fritzing é gratuito e disponível para Linux, Mac OS e Windows. Alguns repositórios Linux tem Fritzing disponível, com fácil instalação :

- [no Manjaro/Arch a última versão está disponível em repositórios da comunidade (AUR)](https://aur.archlinux.org/packages/fritzing/).

### [EasyEDA](https://easyeda.com/)

![image.png](attachment:image.png "EasyEDA")

EasyEDA é outro software gratuito de computador para EDA (Electronic Design Automation), sendo competidor do KiCAD,  bastante elogiado e mais simples de usar que KiCAD.

**EasyEDA tem versão online (web-app)** e offline (para PC desktop), gratuita porém com propaganda, vide [versão gratuita](https://easyeda.com/page/pricing).

Tem [download p/ Mac OS, Linux e Windows](https://easyeda.com/page/download). Alguns Linux tem EasyEDA no repositório, por exemplo em [Linux Arch/Manjaro, no repositório AUR](https://aur.archlinux.org/packages/easyeda-bin/).

- Vbat = 1,5 V (teórico);

Qual escala do multímetro Minipa ET-2042E deve ser usada para ter menor erro em tal medida ?

R.: Escala do voltímetro ?

Vide tabela B do manual do multímetro Minipa ET-2042E, escala de 6V, que tem resolução de 0,001V e precisão de ±(0,5%+2D).

Se em tal escala foi medida a tensão (elétrica) de uma pilha recarregável, $Vbat = 1,329 V$, então calcule a incerteza do Vbat medido, escrevendo a medida no formato (valor nominal +/- incerteza) com o número correto de algarismos significativos.
