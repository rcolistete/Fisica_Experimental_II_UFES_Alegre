### Como ler medidas de tensão $V$ de um voltímetro (ou multímetro nesse modo)

- $V_{bat} = 1,5$ V (teórico);

R.: vide tabela B do manual do multímetro Minipa ET-2042E, **escala de 6V, que tem resolução de 0,001V e precisão de ±(0,5%+2D)**.

Se em tal escala foi medida a tensão (elétrica) de uma pilha recarregável :

$$V_{bat} = 1,329\,\text{V}$$

então calcule a incerteza da $V_{bat}$ medida, escrevendo a medida no formato (valor nominal +/- incerteza) com o número correto de algarismos significativos.

R.: $(1,329 \pm 0,009)$ V

```python
(1.329*0.005)+2*0.001
```
