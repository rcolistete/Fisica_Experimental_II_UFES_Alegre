## Lendo medidas de um Multímetro

Em parte das medidas usaremos o multímetro Minipa [ET-2042E](https://www.minipa.com.br/multimetros/multimetros-digitais/349-et-2042e)



### Como ler medidas de tensão $V$ de um voltímetro (ou multímetro nesse modo)

Para medir :
- $V_{bat} = 1,5$ V (teórico);

qual escala do multímetro Minipa ET-2042E deve ser usada para ter menor erro em tal medida ?

R.: vide tabela B do manual do multímetro Minipa ET-2042E, **escala de 6V, que tem resolução de 0,001V e precisão de $\pm(0,5\%+2D)$**.

Se em tal escala foi medida a tensão (elétrica) de uma pilha recarregável :

$$V_{bat} = 1,329\,\text{V}$$

então calcule a incerteza da $V_{bat}$ medida, escrevendo a medida no formato (valor nominal +/- incerteza) com o número correto de algarismos significativos.

R.: $(1,329 \pm 0,009)$ V

```python
(1.329*0.005)+2*0.001
```

### Como ler medidas de corrente elétrica $i$ de um amperímetro (ou multímetro nesse modo)

Para medir :
- $I_R = 12$  mA (teórico);

qual escala do multímetro Minipa ET-2042E deve ser usada para ter menor erro em tal medida ?

R.: vide tabela C do manual do multímetro Minipa ET-2042E, escala de 60 mA, que tem resolução de 0,01 mA e precisão de $\pm(0,8\%+8D)$.

Se em tal escala foi medida a corrente (elétrica) :

$$I_{R} = 11,58\,\text{mA}$$

então calcule a incerteza da $I_R$, escrevendo a medida no formato (valor nominal +/- incerteza) com o número correto de algarismos significativos.

R.: $(11,58 \pm 0,17)$ mA

```python
(11.58*0.008)+8*0.01
```

### Como ler medidas de resistência elétrica $R$ de um ohmímetro (ou multímetro nesse modo)

Para medir :
- $R = 4,7\,k\Omega$ (valor nominal de um resistor);

qual escala do multímetro Minipa ET-2042E deve ser usada para ter menor erro em tal medida ?

R.: vide tabela ? do manual do multímetro Minipa ET-2042E, escala de $6\,k\Omega$, que tem resolução de $0,001\,k\Omega$ e precisão de $\pm(0,8\% + 3D)$.

Se em tal escala foi medida a resistência (elétrica) :

$$ R = 4,813\,k\Omega $$

então calcule a incerteza de $R$, escrevendo a medida no formato (valor nominal +/- incerteza) com o número correto de algarismos significativos.

R.: $(4,81 \pm 0,04)\,k\Omega$, pois adotamos a dezena limite como sendo 30.

```python
(4.813*0.008) + 3*0.001
```
