# Caracterização de Resistor, Lâmpada Incandescente, Diodo e LED com Multímetros

## Objetivos

Prática de uso de :

- fonte de tensão elétrica regulável;

- protoboard;

- circuito elétrico de uma malha com resistores, lâmpada incandescente, diodo e LED;

- multímetros nos modos voltímetro e amperímetro para caracterização de dispositivo eletrônico via curva $V \times I$.

## Teoria

- resistor, lâmpada incandescente, diodo, LED (diodo emissor de luz) e circuito elétrico de uma malha;

- cálculo de incerteza nas medidas de voltímetro e amperímetro  via consulta da escala, resolução e precisão no manual do multímetro;

- curva $V \times I$ para caracterização de dispositivo eletrônico.

## Equipamento

- fonte de tensão elétrica contínua regulável;

- protoboard;

- cabos, fios e jumpers para protoboard;

- resistores;

- lâmpada incandescente de baixa tensão;

- diodo;

- LED's de cores diferentes;

- 2 multímetros com pontas-de-prova e garras jacaré.

## Procedimento Experimental

Para o relatório :

- fazer registro fotográfico das montagens e das medidas, para algumas constaren no relatório;

- fazer registro por escrito (em papel ou documento eletrônico) de todas as medidas, modelos e escalas dos multímetros;

- fazer diagrama esquemático do circuito eletrônico via algum software de EDA (desenho/projeto) de circuitos eletrônicos.

### Montagem

Fazer a montagem sob orientação do professor :

1) fonte de tensão elétrica contínua regulável com saída conectada no protoboard, **sem alimentação externa e desligada**;

2) **escolher via código de cores** (consulte tabela sobre isso) **os resistores** $R_L$ (limitador) e $R_C$ (carga) indicados e conectá-los em série no protoboard, com um terminal de $R_L$ conectado no terminal positivo da fonte de tensão e um terminal de $R_C$ conectado no terminal negativo da fonte de tensão;

3) usar um multímetro em modo voltímetro, desligado, conectando-o em paralelo ao resistor $R_C$;

4) usar um multímetro em modo amperímetro, desligado, conectando-o em série entre o resistor $R_L$ e o resistor $R_C$;

5) resultando em circuito eletrônico com fonte de tensão elétrica contínua regulável + $R_L$ + amperímetro + $R_C$ em série, e voltímetro em paralelo a $R_C$.

### Medidas

#### Caracterização $V \times I$ de Resistor $R_C$

1) dados os valores dos resistores $R_L$ e $R_C$ escolhidos, calcule a corrente elétrica $I_{max}$ teórica máxima para a tensão $V_{Smax}$ máxima nominal da fonte ("source") de tensão regulável. $R_L = \underline{\ \ \ \ \ \ \ \ \ }\,k\Omega$, $R_C = \underline{\ \ \ \ \ \ \ \ \ }\,k\Omega$, $V_{Smax} = \underline{\ \ \ \ \ \ \ \ \ }\,V$, $I_{max} = \underline{\ \ \ \ \ \ \ \ \ }\,mA$;

2) considerando o divisor de tensão com $R_L$ e $R_C$ e consultando o manual do multímetro, escolher a escala de tensão  elétrica contínua (DC) mais adequada do multímetro em modo voltímetro, i. e., a menor escala capaz de ler a tensão $V_{R_Cmax}$. Escala $= \underline{\ \ \ \ \ \ \ \ \ }\,V$;

3) consultando o manual do multímetro, escolher a escala de corrente elétrica contínua (DC) mais adequada do multímetro em modo amperímetro, i. e., a menor escala capaz de ler a corrente $I_{max}$. Escala $= \underline{\ \ \ \ \ \ \ \ \ }\, mA$;

5) anotar a leitura da tela do multímetro em modo voltímetro, a escala usada, e via consulta do manual do  multímetro, anotar a resolução e precisão de tal escala. Calcular a partir disso a incerteza e expressar o valor da medida com incerteza e o número correto de algarismos significativos. 2) Ao inverter a polaridade da fonte de tensão $V_S$, o que ocorre com as medidas elétricas do resistor $R_C$ ? Então resistor tem polaridade, só conduzindo em um sentido ?
